* Features
1. Cut:

Describes the shape and quality of the diamond’s cut (e.g., Round, Princess). A well-cut diamond reflects light better and has a significant impact on its brilliance and overall value.
2. Color:

Refers to the grade of the diamond’s color, typically ranging from D (colorless) to Z (light yellow or brown). The less color a diamond has, the higher its value.
3. Clarity:

Describes the presence of imperfections (inclusions or blemishes) in a diamond. Clarity grades range from FL (Flawless) to I3 (Included). Higher clarity diamonds, with fewer imperfections, are more valuable.
4. Carat_weight:

Represents the weight of the diamond in carats, which directly impacts its size and price. Larger diamonds are generally more expensive, but other factors (cut, color, clarity) also influence the price.
5. Cut_quality:

A more specific measure of how well the diamond was cut, with grades like Excellent, Very Good, etc. A high-quality cut enhances the diamond’s brilliance and value.
6. Lab:

Refers to the grading laboratory that certified the diamond (e.g., GIA, IGI). Different labs have different reputations, and the credibility of the lab can affect the price.
7. Symmetry:

Measures the precision of the diamond’s shape, which affects how light is reflected. Grades range from Excellent to Poor. Higher symmetry generally leads to better brilliance and higher prices.
8. Polish:

Describes the smoothness of the diamond's surface. A well-polished diamond has better light reflection. Polish grades range from Excellent to Poor.
9. Eye_clean:

A description of whether imperfections are visible to the naked eye. Common values are Yes (imperfections not visible) or No (imperfections are visible).
10. Culet_size:

Refers to the size of the bottom point of the diamond (culet). Most diamonds have a very small or no culet. A large culet may reduce brilliance.
11. Culet_condition:

Describes the condition of the culet, whether it is Good, Poor, etc., and can affect the overall appearance of the diamond.
12. Depth_percent:

A percentage that indicates the height of the diamond (measured from the culet to the table) relative to its width. Ideal depth percentages vary by shape and impact the diamond's sparkle and value.
13. Table_percent:

The size of the flat top of the diamond relative to its overall width. A well-proportioned table size maximizes the diamond's brilliance.
14. Meas_length, Meas_width, Meas_depth:

These measurements give the physical dimensions of the diamond in millimeters, describing its length, width, and depth. They help to determine the shape and proportions of the diamond.
15. Girdle_min and Girdle_max:

These indicate the thickness of the diamond’s girdle (the outer edge), which can affect durability and appearance. Values like Thin, Medium, or Thick describe the range of thickness.
16. Fluor_color and Fluor_intensity:

Fluorescence refers to a diamond’s ability to emit a soft glow when exposed to ultraviolet light. Fluor_color describes the color of this glow (usually blue), and Fluor_intensity indicates its strength (e.g., None, Faint, Strong). Fluorescence can affect price.
17. Fancy_color_dominant_color, Fancy_color_secondary_color, Fancy_color_overtone, Fancy_color_intensity:

These fields are specific to fancy-colored diamonds, which have more intense colors (e.g., pink, blue). These columns describe the dominant and secondary colors, overtone (if applicable), and the intensity of the color (e.g., Vivid, Faint). Fancy-colored diamonds are rarer and can be more valuable.
18. Total_sales_price:

This is the target variable, representing the final sales price of each diamond. It is a critical feature for predictive modeling, as we aim to estimate this value based on the other attributes.

```python
#remove outlires

def remove_outliers_iqr(df, column):
  """
  Removes outliers from a DataFrame column using the IQR method.

  Args:
    df: The pandas DataFrame.
    column: The name of the column to remove outliers from.

  Returns:
    A new DataFrame with outliers removed from the specified column.
  """
  Q1 = df[column].quantile(0.25)
  Q3 = df[column].quantile(0.75)
  IQR = Q3 - Q1
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR
  df_filtered = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
  return df_filtered

# Example usage (replace 'carat_weight' with your desired column)
df_filtered = remove_outliers_iqr(df, 'carat_weight')

# You can also remove outliers from multiple columns using a loop
columns_to_filter = ['carat_weight', 'depth_percent', 'table_percent']  # Add your columns here
for column in columns_to_filter:
  df = remove_outliers_iqr(df, column)
```
